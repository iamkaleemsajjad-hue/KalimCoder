# 04 — Prepare Dataset

Download, clean, and format coding datasets for SFT training.

## Target Datasets
- `codeparrot/github-code`
- `sahil2801/CodeAlpaca-20k`
- Custom curated prompts

```python
from datasets import load_dataset
import json, os

os.makedirs('../datasets/processed', exist_ok=True)
os.makedirs('../datasets/raw', exist_ok=True)
```

```python
# Load CodeAlpaca
ds = load_dataset('sahil2801/CodeAlpaca-20k', split='train')
print(f'CodeAlpaca loaded: {len(ds)} examples')
print(ds[0])
```

```python
# Format into instruction-response pairs
def format_example(example):
    instruction = example.get('instruction', '')
    input_text = example.get('input', '')
    output = example.get('output', '')
    if input_text:
        prompt = f'{instruction}\n\nInput:\n{input_text}'
    else:
        prompt = instruction
    return {'prompt': prompt, 'response': output}

formatted = ds.map(format_example, remove_columns=ds.column_names)
formatted.to_json('../datasets/processed/codealpaca_formatted.jsonl')
print('Saved processed dataset.')
```
