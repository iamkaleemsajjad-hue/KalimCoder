# 02 — Download Model

Download **Qwen3-8B** from Hugging Face Hub to local storage.

```python
import os
from huggingface_hub import login, snapshot_download

# Set your HF token via environment variable HF_TOKEN or paste here
hf_token = os.environ.get('HF_TOKEN', '')
login(token=hf_token)
```

```python
MODEL_ID = 'Qwen/Qwen3-8B'
LOCAL_DIR = '../checkpoints/qwen3-8b-base'

snapshot_download(
    repo_id=MODEL_ID,
    local_dir=LOCAL_DIR,
    ignore_patterns=['*.msgpack', '*.h5', 'flax_model*']
)
print(f'Model downloaded to: {LOCAL_DIR}')
```
