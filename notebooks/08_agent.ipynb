# 08 — Agent

Build a simple coding agent using KaleemCoder as the backbone.

## Capabilities
- Code generation
- Bug fixing
- Code review
- Test generation

```python
from transformers import pipeline

MODEL_PATH = '../adapters/kaleemcoder-sft'

agent_pipeline = pipeline(
    'text-generation',
    model=MODEL_PATH,
    torch_dtype='auto',
    device_map='auto'
)
print('Agent pipeline ready.')
```

```python
SYSTEM_PROMPT = """You are KaleemCoder, an expert software engineering AI assistant.
You help with code generation, debugging, code review, and test writing.
Always provide clean, well-commented, production-ready code."""

def ask_agent(user_query: str, max_new_tokens: int = 512) -> str:
    messages = [
        {'role': 'system', 'content': SYSTEM_PROMPT},
        {'role': 'user', 'content': user_query}
    ]
    result = agent_pipeline(messages, max_new_tokens=max_new_tokens)
    return result[0]['generated_text'][-1]['content']

# Example
response = ask_agent('Write a Python decorator that measures function execution time.')
print(response)
```
